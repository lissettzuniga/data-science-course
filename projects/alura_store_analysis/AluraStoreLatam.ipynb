### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

## Análisis de ingresos por tienda

En este análisis se calcula el ingreso total de cada tienda sumando los valores de la columna **Precio** en cada conjunto de datos.
Esto permite comparar el rendimiento financiero de las tiendas.

```python
ingresos = {
    "Tienda1": tienda["Precio"].sum(),
    "Tienda2": tienda2["Precio"].sum(),
    "Tienda3": tienda3["Precio"].sum(),
    "Tienda4": tienda4["Precio"].sum()
}

print(f"Ingresos Tienda 1: {ingresos['Tienda1']:.0f}")
print(f"Ingresos Tienda 2: {ingresos['Tienda2']:.0f}")
print(f"Ingresos Tienda 3: {ingresos['Tienda3']:.0f}")
print(f"Ingresos Tienda 4: {ingresos['Tienda4']:.0f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(ingresos.keys(), ingresos.values(), color="green")

plt.title("Ingresos totales por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos")

plt.show()
```

## Ventas por categoría

En este análisis se agrupan los datos por **categoría de producto** para identificar cuáles son los tipos de productos más vendidos en cada tienda.

Esto permite comprender mejor las preferencias de los clientes y el rendimiento de cada categoría.

```python
ventas_categoria1 = tienda['Categoría del Producto'].value_counts()
ventas_categoria2 = tienda2['Categoría del Producto'].value_counts()
ventas_categoria3 = tienda3['Categoría del Producto'].value_counts()
ventas_categoria4 = tienda4['Categoría del Producto'].value_counts()

print("Ventas por categoría - Tienda 1")
print(ventas_categoria1)
```

```python
categorias = tienda["Categoría del Producto"].value_counts()

plt.figure(figsize=(7,7))

plt.pie(
    categorias.values,
    labels=categorias.index,
    autopct="%1.1f%%"
)

plt.title("Distribución de ventas por categoría - Tienda 1")

plt.show()
```

```python
print("\nVentas por categoría - Tienda 2")
print(ventas_categoria2)
```

```python
print("\nVentas por categoría - Tienda 3")
print(ventas_categoria3)
```

```python
print("\nVentas por categoría - Tienda 4")
print(ventas_categoria4)
```

```python
categorias_tiendas = pd.concat([
    ventas_categoria1,
    ventas_categoria2,
    ventas_categoria3,
    ventas_categoria4
], axis=1)

categorias_tiendas.columns = ['Tienda 1','Tienda 2','Tienda 3','Tienda 4']
categorias_tiendas
```

## Calificación promedio de cada tienda

En este análisis se calcula la **calificación promedio de los clientes** para cada tienda utilizando la columna `Calificación`.

Esto permite medir el nivel de **satisfacción del cliente** y comparar el desempeño de las tiendas en términos de experiencia de compra.

```python
calificaciones = {
    "Tienda1": tienda["Calificación"].mean(),
    "Tienda2": tienda2["Calificación"].mean(),
    "Tienda3": tienda3["Calificación"].mean(),
    "Tienda4": tienda4["Calificación"].mean()
}

print(f"Calificación Tienda 1: {calificaciones['Tienda1']:.3f}")
print(f"Calificación Tienda 2: {calificaciones['Tienda2']:.3f}")
print(f"Calificación Tienda 3: {calificaciones['Tienda3']:.3f}")
print(f"Calificación Tienda 4: {calificaciones['Tienda4']:.3f}")
```

```python
plt.figure(figsize=(8,5))
plt.bar(calificaciones.keys(), calificaciones.values(), color="orange")

plt.title("Calificación promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación promedio")

plt.show()
```

## Costo de envío promedio por tienda

En este análisis se calcula el costo promedio de envío de cada tienda utilizando la columna **Costo de envío**.

Este indicador permite comprender cuánto pagan en promedio los clientes por el envío de los productos en cada tienda.

```python
envios = {
    "Tienda 1": tienda["Costo de envío"].mean(),
    "Tienda 2": tienda2["Costo de envío"].mean(),
    "Tienda 3": tienda3["Costo de envío"].mean(),
    "Tienda 4": tienda4["Costo de envío"].mean()
}

print(f"Envíos Tienda 1: {envios['Tienda 1']:.3f}")
print(f"Envíos Tienda 2: {envios['Tienda 2']:.3f}")
print(f"Envíos Tienda 3: {envios['Tienda 3']:.3f}")
print(f"Envíos Tienda 4: {envios['Tienda 4']:.3f}")
```

```python
plt.bar(envios.keys(), envios.values())

plt.title("Costo de envío promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Costo promedio de envío")

plt.show()
```

## Informe Final – Análisis de las tiendas Alura Store

### Introducción

El objetivo de este análisis fue evaluar el desempeño de las cuatro tiendas de la cadena Alura Store con el fin de ayudar al Sr. Juan a decidir cuál de ellas debería vender para iniciar un nuevo emprendimiento.  

Para tomar esta decisión se analizaron diferentes indicadores clave, tales como los ingresos totales de cada tienda, las categorías de productos más vendidas, la satisfacción de los clientes a través de las calificaciones, los productos más y menos vendidos y el costo promedio de envío.

Estos indicadores permiten comprender el rendimiento general de cada tienda y detectar posibles debilidades o fortalezas en su operación.

---

### Desarrollo del análisis

**Ingresos totales**

El análisis de ingresos mostró que la Tienda 1 presenta la mayor facturación, seguida por la Tienda 2 y la Tienda 3. Por otro lado, la Tienda 4 presenta el nivel de ingresos más bajo entre las cuatro tiendas, lo que indica un menor volumen de ventas.

**Ventas por categoría**

El análisis de las categorías de productos permitió identificar qué tipos de productos tienen mayor demanda en cada tienda. Algunas categorías concentran gran parte de las ventas, mientras que otras presentan menor rotación, lo cual puede afectar el rendimiento general de la tienda.

**Calificación promedio de los clientes**

Las calificaciones promedio muestran el nivel de satisfacción de los clientes. En general, las tiendas mantienen calificaciones cercanas a valores altos, lo que indica que los clientes están satisfechos con sus compras. Sin embargo, pequeñas diferencias entre tiendas permiten identificar cuáles ofrecen una mejor experiencia al cliente.

**Productos más y menos vendidos**

El análisis de los productos más vendidos muestra cuáles artículos tienen mayor demanda dentro de las tiendas. Asimismo, los productos menos vendidos pueden indicar oportunidades de mejora en la estrategia de ventas o en la oferta de productos.

**Costo de envío promedio**

El análisis del costo de envío permite comprender cuánto pagan en promedio los clientes por el envío de los productos. Este factor puede influir en la decisión de compra, ya que costos de envío más altos pueden afectar la competitividad de una tienda frente a otras.

---

### Conclusión y recomendación

Después de analizar todos los indicadores, se observa que la **Tienda 4 presenta el rendimiento más bajo en comparación con las demás tiendas**. Esta tienda registra los menores ingresos totales, lo que indica un menor volumen de ventas en relación con las otras tiendas de la cadena.

Aunque las calificaciones de los clientes y las ventas por categoría muestran resultados relativamente similares entre las tiendas, el menor nivel de facturación sugiere que la Tienda 4 tiene un desempeño general inferior.

Por lo tanto, **se recomienda que el Sr. Juan venda la Tienda 4**, ya que su rendimiento es el más bajo dentro de la cadena. Esto permitiría concentrar recursos en las tiendas con mejor desempeño y utilizar el capital obtenido para iniciar su nuevo emprendimiento.
