## Misión 1: Equipo de Ventas

Necesitamos construir una visualización para indicar los **10 principales clientes por total de ventas**.

Con ello, el equipo de ventas puede pensar en estrategias para mantener a estos clientes comprando en nuestra tienda.

fecha_pedido: fecha en que el cliente realizó el pedido del producto.

modo_envio: modo de envío del producto especificado por el cliente.

nombre_cliente: nombre del cliente.

segmento_cliente: el segmento al que pertenece el cliente, pudiendo ser B2B o B2C.

ciudad: ciudad de destino del pedido.

provincia: provincia de destino del pedido.

region: región del estado de destino del pedido.

departamento: departamento del producto encargado.

tipo_producto: categoría del producto encargado.

ventas: valor ganado de la venta del producto.

cantidad: cantidad pedida del producto encargado.

ganancia: valor de ganancia/pérdida para la empresa considerando los gastos del producto.

```python
clientes = ventas.groupby(['nombre_cliente'])['ventas'].sum().nlargest(10).copy().reset_index()
```

```python
clientes
```

```python
clientes.columns = ['Clientes','ventas']
clientes['Ranking'] = clientes.index + 1
clientes.set_index('Ranking',inplace=True)
```

```python
clientes
```

```python
estilo = clientes.style
estilo
```

```python
estilo.format({'ventas':'{:,.2f}'})
estilo
```

```python
import pandas as pd

# DataFrame con películas bien valoradas
datos = {
    'Género': ['Drama', 'Comedia', 'Acción', 'Ciencia Ficción', 'Terror'],
    'Nota Media': [4.5, 4.3, 4.2, 4.1, 3.8],
    'Cantidad de Evaluaciones': [1500, 1200, 1000, 800, 600]
}

df_peliculas = pd.DataFrame(datos)

# Añade un ranking de popularidad basado en el número de evaluaciones
df_peliculas['Ranking'] = df_peliculas['Cantidad de Evaluaciones'].rank(method='dense', ascending=False).astype(int)

# Define la columna 'Ranking' como índice del DataFrame
df_peliculas.set_index('Ranking', inplace=True)

# Objeto style
s_pelicula = df_peliculas.style

# Crear un archivo HTML
s_pelicula.to_html('estilo_html.html', index=False)

# Crear un archivo Excel
s_pelicula.to_excel('estilo_xlsx.xlsx', engine='openpyxl')

# Crear un archivo LaTeX
with open('estilo_latex.tex', 'w') as f:
    f.write(s_pelicula.to_latex())
```

# Misión 3: Equipo de Logística

Necesitamos construir una visualización que nos permita entender la **distribución de pedidos por región de Brasil**.

Así, el equipo de logística va a saber dónde hay más oportunidad de enfocar los esfuerzos de material y mano de obra.

```python
region = pd.DataFrame(ventas['region'].value_counts())
region.columns = ['N° Pedidos']
region.index.name = 'Region'
region
```

```python
import numpy as np
```

```python
porcentaje = region['N° Pedidos'].to_numpy()
porcentaje = 100 * porcentaje / porcentaje.sum()
porcentaje
```

```python
region['Porcentaje de pedidos'] = porcentaje
region
```

####**Características visuales que pueden ser alteradas en código mediante la definición de propiedades**
background-color: define el color de fondo de la tabla o de una celda específica;

color: define el color del texto de la tabla o de una celda específica;

border: define el borde de la tabla o de una celda específica, incluyendo el ancho, estilo y color del borde;

border-collapse: define si los bordes de las celdas de la tabla deben estar combinados o separados;

text-align: define la alineación horizontal del texto en una celda específica;

vertical-align: define la alineación vertical del contenido de una celda específica;

font-size: define el tamaño de la fuente del texto de la tabla o de una celda específica;

font-weight: define el grosor de la fuente del texto de la tabla o de una celda específica;

padding: define el espacio interno entre el contenido y el borde de una celda específica.

```python
region_estilo = region.style
region_estilo
```

```python
encabezado = {
    'selector': 'th',
    'props': 'font-weight:bold; font-family:Arial; text-align:right; background-color:white'
}

celdas = {
    'selector': 'td',
    'props': 'background-color:white'
}

region_estilo.set_table_styles([encabezado,celdas])
```

```python
region_estilo.format({'Porcentaje de pedidos':'{:.2f}%'})\
                     .bar(subset='Porcentaje de pedidos',vmin=0,vmax=100,color='#9cd3bb')
```

# Misión 4: Equipo de Logística

Nuevamente el equipo de logística desea replantear su estructura y organización de provisionamiento de productos e, para ello, se pensó en una visualización que permita identificar el estándar de la **cantidad de productos solicitados por mes por departamento**.

```python
meses = ventas.copy()
meses = meses.sort_values('fecha_pedido')
#meses['mes'] = meses['fecha_pedido'].dt.month
meses
```

```python
meses['mes'] = meses['fecha_pedido'].dt.strftime('%Y-%b')
meses = meses.reset_index(drop=True)
```

```python
meses
```

```python
pd.set_option('display.max_columns', None)
```

```python
ventas_mensuales = meses.pivot_table(index=meses.departamento,columns='mes',values='cantidad',aggfunc='sum',sort=False)
ventas_mensuales
```

```python
estilo_mensual = ventas_mensuales.style
estilo_mensual
```

```python
estilo_mensual.set_sticky(axis='index')
```

```python
columnas = {
    'selector': '.col_heading',
    'props': 'font-weight:normal; font-family:Arial'
}

tabla = {
    'selector': 'td,th',
    'props': 'text-align:left'
}

indices = {
    'selector': '.index_name',
    'props': 'font-weight:bold; font-family:Arial; text-align:right'
}

estilo_mensual.set_table_styles([columnas,tabla,indices],overwrite=False)
```

# Misión 5: Informe de Desempeño

El informe de desempeño es un documento muy importante para la empresa. En este, todos los que trabajan en la tienda pueden entender cómo está la empresa. Debido a ello, las tablas bien construídas son esenciales para mostrar los datos de la mejor forma.

Por lo tanto, necesitamos construir una visualización que permita entender la **relación entre tipos de clientes y modo de envío de productos de acuerdo con las ventas**. De esta forma, será posible destacar en el informe cuál es el modo de envío y cuáles son los que más generaron lucro en las ventas.

```python
df_cliente = ventas.pivot_table(index='segmento_cliente',columns='modo_envio',values='ventas',aggfunc='sum')
df_cliente
```

```python
df_cliente['Total'] = df_cliente.sum(axis=1)
df_cliente.loc['Total'] = df_cliente.sum()
df_cliente
```

```python
compra_cliente =df_cliente.style.format('{:,.2f}')
compra_cliente
```

```python
tabla = {
    'selector': 'td,th:not(.index_name)',
    'props': 'font-weight:normal; font-family:Arial; text-align:center; background-color:white'
}

indice = {
    'selector': '.index_name',
    'props': 'font-weight:normal; font-style:italic; text-align:right; color:#696969'
}

compra_cliente.set_table_styles([tabla, indice])
```

```python
compra_cliente.set_table_styles({
    'Total': [
        {
            'selector': 'th',
            'props': 'border-top: 2px solid #181818'
        },
        {
            'selector': 'td',
            'props': 'border-top: 2px solid #181818'
        }
    ],
    'B2B': [
        {
            'selector': 'th',
            'props': 'border-top: 2px solid #181818'
        },
        {
            'selector': 'td',
            'props': 'border-top: 2px solid #181818'
        }
    ]
},overwrite=False,axis=1)
```

```python
compra_cliente.set_table_styles({
    'Total': [{
        'selector': '.true',
        'props': 'background-color: #D8D8D8'
    }]
}, overwrite=False, axis=0)

colores_columna = pd.DataFrame(['false', 'true', 'false'],index=df_cliente['Total'].index,columns=['Total'])

compra_cliente.set_td_classes(colores_columna)
```

```python
import pandas as pd
df = pd.read_csv('https://gist.githubusercontent.com/ahcamachod/52da15c2bbb31e4270b5c267dd051473/raw/62c366878ebd4be2ec899836413c8d1f3e116c01/tienda_libros.csv')
```

```python
# Crear un DataFrame alternativo para tener un nombre de columna coherente con la visualización
df_alteracion = df.rename(columns={'pais': 'Costo por país ($)'})

# Agrupar los datos por tipo de cliente, con la categoría de productos y los países a los que estos clientes importan el producto
df_clientes = df_alteracion.pivot_table(index=['tipo_cliente', 'categoria'], columns = 'Costo por país ($)', values = 'costo_producto', aggfunc = 'sum')

# Renombrar el encabezado para nombres más coherentes
df_clientes.rename_axis(index={'tipo_cliente': 'Tipo cliente', 'categoria': 'Categoría producto'}, inplace = True)

# Agregar una columna que muestre la suma de los costos por tipo de cliente y categoría de producto
df_clientes['Total ($)'] = df_clientes.sum(axis=1)

## Crear la visualización
s_clientes = df_clientes.style.format('{:,.2f}')

# Crear los estilos base para la tabla, desde la estandarización de las fuentes, fondo blanco y alineación del encabezado en la parte superior
tabla = {
    'selector': 'td, th',
    'props': 'font-weight: normal; font-family: Arial; text-align: right; background-color: white;'
}
# Alinear verticalmente el encabezado en la parte superior permite al observador identificar dónde comienza la jerarquización de los grupos B2B y B2C
encabezado = {
    'selector': 'th',
    'props': 'vertical-align: top'
}
s_clientes.set_table_styles([tabla, encabezado])

# Agregar línea de separación en el encabezado superior
# Resaltar las categorías que generaron más costo con la aplicación de negrita en las fuentes
s_clientes.set_table_styles({
    ('B2B', 'BlueRay'): [{'selector': 'th', 'props': 'border-top: 1px solid #181818'},
              {'selector': 'td', 'props': 'border-top: 1px solid #181818'}],
    ('B2B', 'Libro'): [{'selector': 'th', 'props': 'font-weight: bold'}],
    ('B2C', 'Libro'): [{'selector': 'th', 'props': 'font-weight: bold'}],
}, overwrite=False, axis=1)

# Resaltar los elementos en cada tipo de cliente que generaron más costos con una selección
s_clientes.set_table_styles({
    'Total ($)': [{
        'selector': '.true',
        'props': 'font-weight: bold'
    }]
},overwrite=False,axis=0)

colores_columna = pd.DataFrame(['false','false','false','true','false','false','false','true'],index= df_clientes['Total ($)'].index,
                            columns = ['Total ($)'])
# Visualización final
s_clientes.set_td_classes(colores_columna)
```
