#**Visualización de datos con Python: Matplotlib, Seaborn y Plotly**

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

Conocer la tendencia de la inmigración de Colombianos en Canadá

```python
df.set_index('Pais', inplace=True)
df
```

```python
años = list(map(str, range(1980, 2014)))
años
```

```python
colombia = df.loc['Colombia', años]
colombia
```

```python
col_dict = {'Año':colombia.index.to_list(),
            'Inmigrantes':colombia.values.tolist()}
```

```python
col_dict
```

```python
datos_colombia = pd.DataFrame(col_dict)
datos_colombia.head()
```

##**Matplotlib**

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_colombia['Año'],datos_colombia['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

Añadiendo título y etiquetas

```python
plt.figure(figsize=(8,4))
plt.plot(datos_colombia['Año'],datos_colombia['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Inmigración de Colombianos hacia Canadá')
plt.ylabel('Número de inmigrantes')
plt.xlabel('Años')
plt.show()
```

```python
df_comparacion = df.loc[['Brasil', 'Argentina'], años]
df_comparacion
```

```python
df_comparacion = df_comparacion.T
df_comparacion
```

```python
plt.plot(df_comparacion['Brasil'], label='Brasil')
plt.plot(df_comparacion['Argentina'], label='Argentina')
plt.title('Inmigración de Brasil y Argentina a Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend() #agrega una leyenda al gráfico, reconociendo el parámetro label
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_colombia['Año'],datos_colombia['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de Colombianos hacia Canadá\n1980-2013')
ax.set_xlabel('Año')
ax.set_ylabel('Inmigrantes')
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(10,3))
axs[0].plot(datos_colombia['Año'],datos_colombia['Inmigrantes'])
axs[0].set_title('Inmigración de Colombianos hacia Canadá\n1980-2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))

axs[1].boxplot(datos_colombia['Inmigrantes'])
axs[1].set_title('Boxplot de la inmigración de Colombianos \nhacia Canadá 1980-2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Número de inmigrantes')

plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle('Comparación de países')

axs[0,0].plot(df.loc['Colombia',años])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil',años])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina',años])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú',años])
axs[1,1].set_title('Perú')

ymin =0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Número de Inmigrantes')
  ax.grid()

plt.show()
```

```python
fig, axs = plt.subplots(figsize=(8,3))
axs.plot(datos_colombia['Año'],datos_colombia['Inmigrantes'],lw=3) #marker='o')
axs.xaxis.set_major_locator(plt.MultipleLocator(5))
axs.set_title('Inmigración de Colombianos hacia Canadá\n1980-2013',fontsize=18,loc='left')
axs.set_xlabel('Año',fontsize=14)
axs.set_ylabel('Inmigrantes',fontsize=14)
axs.yaxis.set_tick_params(labelsize=12)
axs.xaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
sudamerica = df.query('Region=="América del Sur"')
sudamerica.head()
```

```python
colores = [
    'royalblue', 'orange', 'forestgreen', 'orchid',
    'purple', 'brown', 'slateblue', 'gray', 'olive',
    'navy', 'teal', 'tomato'
]
```

```python
fig,ax = plt.subplots(figsize=(10,3))
ax.barh(sudamerica.index,sudamerica['Total'],color=colores)
ax.set_title('Inmigración de Sudamericanos hacia Canadá\ndurante el periodo de 1980-2013',fontsize=18,loc='left')
ax.set_ylabel('')
ax.set_xlabel('Inmigrantes',fontsize=14)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
sudamerica_sorted = sudamerica.sort_values('Total',ascending=True)
fig,ax = plt.subplots(figsize=(10,3))
ax.barh(sudamerica_sorted.index,sudamerica_sorted['Total'],color=colores)
ax.set_title('Inmigración de Sudamericanos hacia Canadá\ndurante el periodo de 1980-2013',fontsize=18,loc='left')
ax.set_ylabel('')
ax.set_xlabel('Inmigrantes',fontsize=14)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
colores=[]
for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    colores.append('forestgreen')
  else:
    colores.append('silver')

fig,ax = plt.subplots(figsize=(10,3))
ax.set_frame_on(False)
ax.xaxis.set_visible(False)
ax.tick_params(axis='y',which='both',size=0)

ax.barh(sudamerica_sorted.index,sudamerica_sorted['Total'],color=colores)
ax.set_title('Inmigración de Colombianos hacia Canadá\ndurante el periodo de 1980-2013',fontsize=18,loc='left')
ax.set_ylabel('')
ax.yaxis.set_tick_params(labelsize=12)
#ax.set_xlabel('Inmigrantes',fontsize=14)
#ax.xaxis.set_tick_params(labelsize=12)


for i,j in enumerate(sudamerica_sorted['Total']):
  ax.text(j+20,i,str(j),color='black',fontsize=8.5, ha='left',va='center')

fig.savefig('/content/inmigrantes_sudamericanos_canada.png',
         transparent=False,dpi=300,bbox_inches='tight')

plt.show()
```

##**Seaborn**

```python
import seaborn as sns
```

```python
sns.set_theme()
```

```python
top_10 = df.sort_values('Total',ascending=False).head(10)
top_10
```

```python
from numpy._core.defchararray import title
ax = sns.barplot(data=top_10,x='Total',y=top_10.index,orient='h')
ax.set(title='Top 10 paises con mayor migración a Canadá\n1980-2013',
       xlabel='Número de inmigrantes',
       ylabel='Año')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax = sns.barplot(data=top_10,x='Total',y=top_10.index,orient='h')
ax.set_title('Top 10 paises con mayor migración a Canadá\n1980-2013', loc='left',fontsize=16)
ax.set_xlabel('Número de inmigrantes',fontsize=14)
ax.set_ylabel('')

plt.show()
```

```python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8, 4))
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  ax.tick_params(axis='y',which='both',size=0)

  ax = sns.barplot(data=top_10,x='Total',y=top_10.index,hue=top_10.index,orient='h',palette=palette,
                   legend=False)
  ax.set_title('Top 10 paises con mayor migración a Canadá\n1980-2013', loc='left',fontsize=16)
  ax.set_xlabel('Número de inmigrantes',fontsize=14)
  ax.set_ylabel('')
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black',fontsize=8.5, ha='left',va='center')

  plt.show()
```

##**Plotly**

```python
import plotly.express as px
```

```python
fig = px.line(datos_colombia, x='Año',y='Inmigrantes',
              title='Inmigración de Colombianos hacia Canadá 1980-2013')
fig.update_traces(line_color='green',line_width=3)

fig.update_layout(width=800,height=400,xaxis={'tickangle':-45},
                  xaxis_title='Año',yaxis_title='Numero de inmigrantes',
                  font_family='Arial',font_size=14,font_color='grey')
fig.show()
```

```python
sudamerica.drop(['Continente','Region','Total'],axis=1,inplace=True)
sudamerica
```

```python
sudamerica = sudamerica.T
sudamerica
```

```python
fig = px.line(sudamerica, x=sudamerica.index,y=sudamerica.columns,
              title='Inmigración de s hacia Canadá 1980-2013',
              color='Pais',markers=True)

fig.update_layout(width=800,height=400,xaxis={'tickangle':-45},
                  xaxis_title='Año',yaxis_title='Numero de inmigrantes',
                  font_family='Arial',font_size=14,font_color='grey')
fig.write_html('/content/inmigracuon_sudamericanos_canada.html')

fig.show()
```
